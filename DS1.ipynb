```python
import matplotlib.pyplot as plt

# Sample gender data
genders = ['Male', 'Female', 'Male', 'Male', 'Female', 'Female', 'Male', 'Male', 'Female', 'Female']

# Counting the frequency of each gender
gender_counts = {gender: genders.count(gender) for gender in set(genders)}

# Plotting the bar chart
plt.bar(gender_counts.keys(), gender_counts.values(), color=['r', 'y'])
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.title('Distribution of Genders in a Population')
plt.show()
```
